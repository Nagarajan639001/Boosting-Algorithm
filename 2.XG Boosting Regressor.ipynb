```python
# Open the dataset
```

```python
dataset= pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Remove the catagorical data (Converting str to Numbers)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#Splite input & Out put data set
```

```python
dataset.columns
```

```python
independent=dataset[['age','bmi','children','sex_male','smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
```

```python
dependent
```

```python
# split the Train set & test set dataset
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
# Model Improve for using the Standardisation
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

```python
# Model creation-Train set
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regressor= GradientBoostingRegressor(loss='squared_error',n_estimators=100,random_state=0)
regressor.fit(X_train,Y_train)
```

```python
# Evaluation Method Presidcting the Test set result
```

```python
Y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score =r2_score(Y_test,Y_pred)
```
